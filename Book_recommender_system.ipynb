```python
ratings=pd.read_csv("BX-Book-Ratings.csv",sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
ratings.head(2)
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'book-rating'},inplace=True)
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
bk=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books=bk[['user_id','ISBN','book-rating','Book-Title','Book-Author','Year-Of-Publication','Publisher']]
```

```python
ratings_with_books.head(2)
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('Book-Title')['book-rating'].count().reset_index()
```

```python
number_rating
```

```python
number_rating.rename(columns={'book-rating':'Number of ratings'},inplace=True)
```

```python
number_rating.head(2)
```

```python
final_rating=ratings_with_books.merge(number_rating,on='Book-Title')
```

```python
final_rating.head(2)
```

```python
 final_rating.shape
```

```python
final_rating=final_rating[final_rating['Number of ratings']>=50]
```

```python
final_rating.drop_duplicates(['user_id','Book-Title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='Book-Title',values='book-rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
print(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
   print(book_pivot.index[suggestions[i]])
```

```python
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
  for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
print(recommend_book('Harry Potter and the Chamber of Secrets (Book 2)'))
```
